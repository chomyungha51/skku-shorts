```python
#A18
total = 0
count = 0
menu_list = {"아메리카노": 2000,"카페라떼": 2500,"밀크티": 3000, "베이글":3000,"모닝세트":3500  }

while(True):
  menu = input("원하는 메뉴를 입력하세요. 주문완료시 '완료'를 입력하세요: ")
  if(menu == "완료"):
    print("총", count,"가지 메뉴 주문하셨으며, 총 금액은", total, "입니다")
    print("감사합니다!")
    break
  else:
    print(menu+"를 입력하셨습니다\n")
    total += menu_list[menu]
    count += 1
```

```python
#A25
menu = {'아메리카노':2000, '카페라떼':2500, '밀크티':3000, '카푸치노':3500, 
        '아이스티':2500, '연유라떼':2500, '바닐라라떼':3500, '카페모카':3500,
        '카라멜마끼아또':4000, '레몬에이드':3000, '자몽에이드':3000}
def order(*args):
  for i in args:
    if i in menu:
      print(i,"는 ",menu[i],"원입니다.", sep="")
    else:
      print(i+"는 판매하지 않습니다.")
order('아메리카노','카페라떼','아포가토','카푸치노')

order('아이스티','자몽에이드')
```
